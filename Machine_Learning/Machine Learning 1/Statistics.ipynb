```python
import pandas as pd

df = pd.DataFrame({'Name': ['Hinat', 'thuki', 'suki', 'momo'],
                   'Salary': [5,2,3,5]})
print(df)
```

```python
# calculate mean, median, mode, min, max
print('min', df['Salary'].min())
print('max', df['Salary'].max())
print('mean', df['Salary'].mean())
print('median', df['Salary'].median())
print('mode', df['Salary'].mode())
print('range', df['Salary'].max()-df['Salary'].min())
```

```python
df['Salary'].describe()
```

```python
# histogram of salaries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salary = df['Salary']
salary.plot.hist(title = 'Salary Distribution', color = 'grey', bins = 25)
plt.axvline(salary.mean(), color = 'violet', linestyle = 'dashed', 
            linewidth = 2)
plt.axvline(salary.median(), color = 'red', linestyle = 'dashed', 
            linewidth = 2)
plt.show()
```
