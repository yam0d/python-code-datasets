# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
plt.plot(x,y)
```

```python
plt.scatter(x[::10],y[::10], color = 'red')
```

```python
plt.hist(x)
```

```python
plt.plot(x,y, color = 'blue')
plt.plot(x,np.cos(x), '--g')
```
